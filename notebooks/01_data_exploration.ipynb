## 1.6 Exploratory Data Analysis (EDA)
### This code checks dataset

```python
# List all class folders in dataset
classes = sorted(os.listdir(BASE_INPUT_DIR))
# Keep real folders
classes = [c for c in classes if os.path.isdir(os.path.join(BASE_INPUT_DIR, c))]
print(f"Total number of classes (diseases/healthy): {len(classes)}")
print("\nFirst 5 classes:")
# Show first classes
for i, class_name in enumerate(classes[:5]):
    print(f"  {i+1}. {class_name}")
# Count the number of images by class
class_counts = {}
total_images = 0
print("\nImage count per class:")
for class_name in classes:
    class_path = os.path.join(BASE_INPUT_DIR, class_name)
    num_images = len(os.listdir(class_path))
    class_counts[class_name] = num_images
    total_images += num_images
    print(f"  {class_name}: {num_images} images")
print(f"\nTotal images in the dataset: {total_images}")
# Create df to visualize the distribution
df_counts = pd.DataFrame(class_counts.items(), columns=['Clase', 'Conteo'])
df_counts = df_counts.sort_values(by='Conteo', ascending=False)
plt.figure(figsize=(15, 6))
plt.bar(df_counts['Clase'], df_counts['Conteo'])
plt.xticks(rotation=90, fontsize=8)
plt.title('Distribution of Images per Class (PlantVillage Dataset)')
plt.show()
```
