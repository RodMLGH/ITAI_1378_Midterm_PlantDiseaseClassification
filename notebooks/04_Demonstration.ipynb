###   5.4 Prediction Function (Real-world Use)
### This code define prediction function.

```python
# Prediction function definition
def predict_image(image_path, model, class_names):
    # Load and preprocess the image
    image = Image.open(image_path).convert('RGB')
    transform = val_test_transforms
    # Adds batch dimension
    image_tensor = transform(image).unsqueeze(0)
    image_tensor = image_tensor.to(device)
    # Perform prediction
    with torch.no_grad():
        output = model(image_tensor)
    # Calculate probability using Softmax
    probabilities = torch.nn.functional.softmax(output[0], dim=0)
    # Get predicted class and confidence
    conf_score, top_class_index = torch.max(probabilities, 0)
    # Convert tensors to Python values
    predicted_class = class_names[top_class_index.item()]
    confidence_score = conf_score.item() * 100
    # Display result
    print("--- Prediction ---")
    print(f"Predicted Class: {predicted_class}")
    print(f"Confidence Score: {confidence_score:.2f}%")
    return predicted_class, confidence_score
```
